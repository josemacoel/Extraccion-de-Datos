### **Filtro CVE_CLPV (Valores de 1000 a 2000)**

```python
df['CVE_CLPV'] = df['CVE_CLPV'].replace('[^0-9]+', '0', regex=True)
df['CVE_CLPV'] = df['CVE_CLPV'].astype(int)
filtro1=df.loc[(df["CVE_CLPV"] >= 1000) & (df['CVE_CLPV'] <= 2000)]
filtro1
```

### **Filtro CVE_VEND (Todas las claves excepto “5” y “4”)**

```python
filtro2 = df.loc[(df['CVE_VEND'] != 4) & (df['CVE_VEND'] != 5)]
filtro2
```
